```python
df3 = pd.DataFrame({'customer':['101','102','103','104'],
                    'category':['cat2','cat2','cat1','cat3'],
                   'important':['yes','no','yes','yes'],
                   'sales':[123,52,214,663]})
df3
```

```python
df4 = pd.DataFrame({'customer':['100','103','104','105'],
                    'color':['yellow','green','green','blue'],
                   'distance':[12,9,44,21],
                   'sales':[123,214,663,331]})
df4
```

```python
pd.concat([df3,df4],axis=0,sort=False)
```

```python
pd.concat([df3,df4],axis=0,sort=True)
```
